```python
import os
import pandas as pd
from collections import defaultdict


# Define the main directory where subfolders are located
MSAD_result_path = '/mnt/c/Arbeid/Github_Repo/MSAD_work_sync/tmp/MSAD_work/results/votes2graph/in_distribution/un-supervised/'  # supervised
MSAD_choices_by_model = defaultdict(list)

# Recursively go through all the subfolders
for root, dirs, files in os.walk(MSAD_result_path):
    for file in files:
        if file.endswith('.csv') and '-' in file:
            # Get the full path of the CSV file and extract model name
            file_path = os.path.join(root, file)
            model_name = file.split('-')[0]
            
            # Extract the model name from the folder name (last folder in the path)
            feature_name = os.path.basename(root)
            
            # Read the CSV file into a DataFrame
            df = pd.read_csv(file_path, index_col=1)            
            df.drop(columns=["Unnamed: 0"], inplace=True)
            df.rename_axis("", inplace=True)
            df.rename(columns={'choice': feature_name + '-' + model_name + '_' + str(len(MSAD_choices_by_model[feature_name + '-' + model_name]))}, inplace=True)
            #df.rename(columns={'choice': feature_name + '-' + '_'.join(model_name.split('_')[:-1]) + '_' + str(len(MSAD_choices_by_model[feature_name + '-' + model_name]))}, inplace=True)
            
            # to number the runs
            MSAD_choices_by_model[feature_name + '-' + model_name].append(df)
            

```

```python
def update_acc(row, column):
    column_index = int(row[column])
    if column_index < 0 or column_index >= len(df_total.columns):
        return None  # or some default value, e.g., np.nan
    column_name = df_total.columns[column_index]
    value = row[column_name]
    return value


df_total_avg_acc = dict()
for model in MSAD_choices_by_model.keys():
    df_total_avg_acc[model] = pd.concat(MSAD_choices_by_model[model], axis=1, join='inner', ignore_index=False)

    # Join with df_total
    df_total_avg_acc[model] = df_total.join(df_total_avg_acc[model], how='inner')

    #print(f"Columns for model {model}: {df_total_avg_acc[model].columns}", df_total_avg_acc[model])  # Debugging print

    # Update each column
    for col in df_total_avg_acc[model].columns:
        if model not in col:
            continue
        df_total_avg_acc[model][col] = df_total_avg_acc[model].apply(lambda row: update_acc(row, col), axis=1)

    # Filter and calculate mean
    filtered_df = df_total_avg_acc[model].filter(like=model)
    #print(filtered_df)  # Debugging print
    df_total_avg_acc[model] = filtered_df.mean(axis=1)

    #print(model, df_total_avg_acc[model])  # Final result for the model
```

```python
def beautify_method_name(method_name):
    formal_method_names = {
        'Random Forest': 'Random Forest',
        'Mlp': 'MLP',
        'Knn': 'KNN',
        'Ada Boost': 'AdaBoost',
        'Qda': 'QDA',
        'Decision Tree': 'Decision Tree',
        'Svc Linear': 'SVC Linear',
        'Bayes': 'Bayes',
        'Sit Conv Patch': 'SiT-conv',
        'Sit Stem Original': 'SiT-stem',
        'Sit Linear Path': 'SiT-linear',
        'Sit Stem Relu': 'SiT-stem-ReLU',
        'Resnet Default': 'ResNet',
        'Convnet Default': 'ConvNet',       
        'Sit Linear Patch': 'SiT-Linear',
        'Inception Time Default': 'Inception Time'
    }
    
    try:
        method_name = col_name.split('-')[1]
    except:
        pass
    method_name = [word.capitalize() for word in method_name.split('_')]
    window_length = method_name[-1]
    method_name = ' '.join(method_name[:-1])
    #print(method_name)
    method_name = method_name.replace(method_name, formal_method_names[method_name])
    method_name += ' ' + str(window_length)
    
    return method_name

def beautify_feature_model(feature_name):
    formal_feature_names = {'catch22': 'catch22',
                            'tsfresh': 'TSFresh'}
    
    return feature_name.replace(feature_name, formal_feature_names[feature_name])
```

```python
from sklearn.model_selection import train_test_split
# Define validation set size
validation_size = 0.15
num_samples = len(df)
val_samples = int(num_samples * validation_size)

# Get counts of each class to balance
class_counts = df['label'].value_counts()
min_class_val_size = int(val_samples / len(class_counts))  # Equal number from each class

# Collect indices for the validation set
val_indices = []
for label in class_counts.index:
    indices = df[df['label'] == label].index.tolist()
    val_indices.extend(indices[:min_class_val_size])

# Remaining indices are for training
train_indices = list(set(df.index) - set(val_indices))

# Split the data
train_df = df.loc[train_indices]
val_df = df.loc[val_indices]
```
